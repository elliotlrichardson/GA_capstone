# Cleaning and EDA: UCLA Data

In this notebook, I cleaned and formatted the Pew dataset so it was ready for modeling. I also did some light EDA to see how the features interacted with the target variable. 

### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data/ucla.txt')
print(df.shape)
df.head()
```

```python
df.drop(columns = df.columns[df.isnull().sum()>6000], inplace = True)
df.drop(columns = df.columns[df.columns.str.startswith('primary_')], inplace = True)
df.drop(columns = 'foreign_born')
df.head()
```

## Correcting datatypes

A few of the columns have some empty strings in them so the entire column is cast as string objects. So I'm going to try to go through and replace all of those with nulls so I can account for them later.

```python
counts = []

for col in df.columns[df.dtypes == object]:
    count = 0
    
    for val in ['888','999',' ', '']:
        try:
            count += df[col].value_counts()[val]
        except:
            pass
        
    counts.append(count)
```

```python
null_tuples = list(zip(df.columns[df.dtypes == object], counts))

cols_w_nulls = []

for col, count in null_tuples:
    if count > 1000:
        cols_w_nulls.append(col)

other_cols = [col for col in df.columns if col not in cols_w_nulls]
```

```python
df.drop(columns = cols_w_nulls, inplace=True)
df.drop(columns = df.columns[df.columns.str.startswith('extra')],inplace=True)
```

```python
map_dict = {}

for i in range(1,888):
    map_dict[i] = i
    map_dict[str(i)] = i
    
map_dict[' '] = np.nan
map_dict['888'] = np.nan
map_dict['999'] = np.nan
map_dict[''] = np.nan
```

```python
for col in df.columns[df.dtypes == object][:-2]:
    df[col] = df[col].map(map_dict)
    
for col in df.columns[df.dtypes != object]:
    df[col].replace(999, np.nan)
    df[col].replace(888, np.nan)
```

```python
df.drop(columns = ['response_id','start_date'], inplace = True)
df.drop(columns = df.columns[df.isnull().sum() > 500], inplace = True)
df.dropna(inplace=True)
```

```python
for col in df.columns[df.dtypes != object]:
    df[col].replace(999, np.nan, inplace=True)
    df[col].replace(888, np.nan, inplace=True)
```

```python
cols = [col for col in df.columns[df.isnull().sum() > 150] if col != 'group_favorability_the_police']
df.drop(columns = cols, inplace=True)
df.dropna(inplace=True)
```

```python
df['group_favorability_the_police'] = 5 - df['group_favorability_the_police']
df.head()
```

## Mapping

Now that I've dealt with those null values, I'm going to re-map some of the values in the remaining columns. For example, I'm going to map binary columns to 0s and 1s instead of 1s and 2s since that's easier for me to interpret. I'm also going to re-map categorical variables to strings so that the dummying process produces more interpretable column names.

```python
binary_cols = [col for col in df.columns if set(df[col].unique()) == {1,2}]

for col in binary_cols:
    df[col] = df[col].map({1:1,2:0})
```

```python
employment_dict = {1: 'employed', 2: 'homemaker', 3: 'retired', 4: 'unemployed',
                   5: 'part_time', 6: 'disabled', 7: 'student', 8: 'self_employed',
                   9: '1other'}

language_dict = {1: 'spanish', 2: '1other', 3: 'english'}

religion_dict = {1: 'protestant', 2: 'catholic', 3: 'mormon', 4: 'orthodox',
                 5: 'christian', 6: 'jewish', 7: 'muslim', 8: 'buddhist',
                 9: 'hindu', 10: 'atheist', 11: 'agnostic', 12: 'none',
                 13: '1other'}

orientation_dict = {1: 'heterosexual', 2: 'gay_man', 3: 'lesbian', 4: 'bisexual',
                   5: '1refused', 6: 'other'}

union_dict = {1: 'yes', 2: '1former_member', 3: 'never'}

gun_dict = {1: 'yes', 2: '1other_member', 3: 'no'}

region_dict = {1: 'northeast', 2: 'midwest', 3: 'south', 4: 'west'}

race_dict = {1: 'white', 2: 'black', 3: 'native', 4: 'asian', 5: 'asian',
             6: 'asian', 7: 'asian', 8: 'asian', 8: 'asian',
             9: 'pacific_islander', 10: 'pacific_islander', 
             11: 'pacific_islander', 12: 'pacific_islander',
             13: 'pacific_islander', 14: 'pacific_islander', 15: '1other'}

df['employment'] = df['employment'].map(employment_dict)
df['language'] = df['language'].map(language_dict)
df['religion'] = df['religion'].map(religion_dict)
df.rename(columns = {'is_evangelical': 'religion_evangelical'}, inplace=True)
df['orientation_group'] = df['orientation_group'].map(orientation_dict)
df.rename(columns = {'orientation_group': 'orientation'}, inplace=True)
df['in_union'] = df['in_union'].map(union_dict)
df['household_gun_owner'] = df['household_gun_owner'].map(gun_dict)
df['census_region'] = df['census_region'].map(region_dict)
df.rename(columns = {'census_region': 'region'},inplace=True)
df['race_ethnicity'] = df['race_ethnicity'].map(race_dict)
```

```python
final_cols = ['group_favorability_the_police', 'news_sources_facebook','news_sources_cnn','news_sources_msnbc',
              'news_sources_fox', 'news_sources_network', 'news_sources_localtv','news_sources_telemundo',
              'news_sources_npr', 'news_sources_amtalk','news_sources_new_york_times','news_sources_local_newspaper',
              'news_sources_other', 'employment', 'language','religion', 'religion_evangelical',
              'orientation', 'in_union', 'household_gun_owner','fc_smallgov','age', 'gender', 'region',
              'race_ethnicity','household_income', 'education', 'state']
```

```python
df = df.loc[:,final_cols]

dummy_cols = ['state','employment','language','religion','orientation','in_union',
              'race_ethnicity','household_gun_owner']

df = pd.get_dummies(df, columns = dummy_cols, drop_first = True)
df = pd.get_dummies(df, columns = ['region'])
df.head()
```

## Charting relationships

I'm not sure what to expect as far as relationships with the target variable, so I'm going to create a loop to chart a ton of the relationships in one go. They won't be super interpretable without tick labels, but they'll help me get a sense of which features and specific values have relationships with support for the police.

```python
def grouped_chart(cat,sort = False):
    
    cols = df.columns[df.columns.str.startswith(cat)]
    if len(cols) > 1:

        grouped = df.groupby('group_favorability_the_police').sum()[df.columns[df.columns.str.startswith(cat)]]
        for col in df.columns[df.columns.str.startswith(cat)]:
            grouped[col] = grouped[col] / df[col].sum()
            
        grouped = grouped.T.sort_values(4)
        grouped.sort_index(inplace=sort)
        ticks = [' '.join([word.capitalize() for word in col[len(cat)+1:].split('_')]) for col in list(grouped.index)]
    
    else:
        col = cols[0]
        vals = list(df[col].unique())
        grouped = pd.DataFrame(index = range(1,5))

        for val in vals:
            grouped[val] = [0] * grouped.shape[0]
            total = df[df[col] == val].shape[0]
            for i in range(1,5):
                count = df[df[col]==val].groupby('group_favorability_the_police').count().mean(axis=1)[i]

                grouped.loc[i,val] = count
                
            grouped[val] = grouped[val] / total
        grouped = grouped.T.sort_values(4)
        grouped.sort_index(inplace=sort)
        ticks = grouped.index

    
    fields = grouped.columns
    labels = ['Very unfavorable', 'Somewhat unfavorable', 'Somewhat favorable','Very favorable']
    colors = ['firebrick','lightcoral','lightgreen','forestgreen']
    
    if len(cols) > 6:
        rotate = 45
    else:
        rotate = 0
    
    plt.figure(figsize = (16,10))
    bottom = len(grouped) * [0]
    
    for i, name in enumerate(fields):
        plt.bar(grouped.index, grouped[name], bottom = bottom, color = colors[i], alpha = 0.8)
        bottom = bottom + grouped[name]
    
    title = ' '.join([word.capitalize() for word in cat.split('_')])
    
    plt.title('Support for the Police by '+title,fontsize = 24)
    plt.legend(labels, ncol=len(fields),loc='upper center',frameon=False,fontsize = 14)
    plt.ylabel('Percent of Respondants',fontsize = 18)
    plt.ylim(bottom = 0, top = 1.1)
    plt.yticks(np.arange(0,1.1,0.1), [str(i)+'%' for i in range(0,101,10)],fontsize = 14)
    plt.xticks(grouped.index,labels = ticks, fontsize = 16, rotation = rotate)
    plt.show()

    
groups = ['employment','language','religion','orientation', 'in_union','household_gun_owner',
          'region','race_ethnicity','gender']

for group in groups:
    grouped_chart(group)
    
grouped_chart('household_income',sort = True)
```
